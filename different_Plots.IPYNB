```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
from matplotlib import pyplot as plt
```

Datensatz für Model 3: Verteilung der Werte für einige Untersuchungsarten 

```python
df = pd.read_csv('model5_ausmodel4.csv')
df = df.iloc[:, 1:]
tot = pd.read_csv('Verstorben.csv')
tot = tot.iloc[:, 1:]

uniquetote = tot['Pseudonym'].unique()
df1 = df[df['Pseudonym'].isin(uniquetote)] #DF nur mit den Toten
df2 = df[~df['Pseudonym'].isin(uniquetote)] #DF nur mit den Lebenden
```
